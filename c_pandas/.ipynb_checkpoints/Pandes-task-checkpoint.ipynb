```python
gdp_by_region = happyness_df.groupby('region')['gdp_per_capita'].agg(['mean'])
print(gdp_by_region)
# print('=' *50)
# gdp_by_region_sorted = gdp_by_region.sort_values(by='mean', ascending=False)
# print(gdp_by_region_sorted)
```

```python
social_by_region = happyness_df.groupby('region')['social_support'].agg(['mean'])
print(social_by_region)
# print('=' *50)
# social_region_sorted = social_by_region.sort_values(by='mean', ascending=False)
# print(social_region_sorted)
```

```python
healthy_by_region = happyness_df.groupby('region')['healthy_life_expectancy'].agg(['mean'])
print(healthy_by_region)
# print('=' *50)
# healthy_by_region = healthy_by_region.sort_values(by='mean', ascending=False)
# print(healthy_by_region)
```

```python
free_by_region = happyness_df.groupby('region')['freedom_to_make_life_choices'].agg(['mean'])
print(free_by_region)
# print('=' *50)
# free_by_region = free_by_region.sort_values(by='mean', ascending=False)
# print(free_by_region)
```

```python
generosity_by_region = happyness_df.groupby('region')['generosity'].agg(['mean'])
print(generosity_by_region)
# print('=' *50)
# generosity_by_region = generosity_by_region.sort_values(by='mean', ascending=False)
# print(generosity_by_region)
```

```python
perceptions_by_region = happyness_df.groupby('region')['perceptions_of_corruption'].agg(['mean'])
print(perceptions_by_region)
# print('=' *50)
# perceptions_by_region = perceptions_by_region.sort_values(by='mean', ascending=False)
# print(perceptions_by_region)
```

```python
merged_data = pd.concat([happiness_by_region, gdp_by_region, social_by_region, healthy_by_region, free_by_region, generosity_by_region, perceptions_by_region], axis=1)
merged_data.columns = ['happy', 'gdp', 'social', 'healthy', 'freedom', 'generosity', 'perceptions']
merged_data

merged_data_sorted = merged_data.sort_values(by='happy', ascending=False)
display(merged_data_sorted)
```

```python
import matplotlib.pyplot as plt

# ê·¸ëž˜í”„ ìƒì„±
plt.figure(figsize=(10, 6))

# ê° ì—´ì— ëŒ€í•œ ì„  ê·¸ëž˜í”„ ìƒì„±
for column in merged_data.columns:
    plt.plot(merged_data.index, merged_data[column], marker='o', label=column)

# ê·¸ëž˜í”„ ì œëª© ì„¤ì •
plt.title('Happiness Score by Various Indicators')

# xì¶• ë ˆì´ë¸” ì„¤ì •
plt.xlabel('Region')

# yì¶• ë ˆì´ë¸” ì„¤ì •
plt.ylabel('Value')

# ë²”ë¡€ ì¶”ê°€
plt.legend()

# xì¶• ë¼ë²¨ íšŒì „
plt.xticks(rotation=45)

# ê·¸ëž˜í”„ ì¶œë ¥
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

happyness_df =  pd.read_csv('./datasets/happiness_report_2023.csv')
happyness_df
```

```python
# happyness_df.loc[happyness_df['region'] == 'Afghanistan', 'region'] = 'Africa-out'
happyness_df['region'].value_counts()
```

```python
hp_mean_df = happyness_df.groupby('region').mean('happiness_score')
merged_data_sorted = hp_mean_df.sort_values(by='happiness_score', ascending=False)
display(merged_data_sorted)
```

```python
from scipy.stats import iqr
Asia = happyness_df.groupby('region').get_group('Asia')

happiness_Q1 = Asia.describe().T.loc['happiness_score', '25%']
happiness_Q3 = Asia.describe().T.loc['happiness_score', '75%']
print(happiness_Q1, happiness_Q3)

iqr_value = iqr(Asia.happiness_score)
print(iqr_value)
lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'ì •ìƒì¹˜ ë²”ìœ„: {lower_bound} ~ {upper_bound}')

outliers = Asia[(Asia['happiness_score'] < lower_bound) | (Asia['happiness_score'] > upper_bound)]
outliers
```
