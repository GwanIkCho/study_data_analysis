```python
import pandas as pd

happyness_df =  pd.read_csv('./datasets/happiness_report_2023.csv')
happyness_df
```

```python
happyness_df.loc[happyness_df['region'] == 'Central and Eastern Europe', 'region'] = 'Europe'
happyness_df.loc[happyness_df['region'] == 'Western Europe', 'region'] = 'Europe'
happyness_df.loc[happyness_df['region'] == 'Commonwealth of Independent States', 'region'] = 'Asia'
happyness_df.loc[happyness_df['region'] == 'East Asia', 'region'] = 'Asia'
happyness_df.loc[happyness_df['region'] == 'South Asia', 'region'] = 'Asia'
happyness_df.loc[happyness_df['region'] == 'Southeast Asia', 'region'] = 'Asia'
happyness_df.loc[happyness_df['region'] == 'Latin America and Caribbean', 'region'] = 'America'
happyness_df.loc[happyness_df['region'] == 'North America and ANZ', 'region'] = 'America'
happyness_df.loc[happyness_df['region'] == 'Middle East and North Africa', 'region'] = 'Africa'
happyness_df.loc[happyness_df['region'] == 'Sub-Saharan Africa', 'region'] = 'Africa'
happyness_df['region'].value_counts()
```

```python
hp_mean_df = happyness_df.groupby('region').mean('happiness_score')
merged_data_sorted = hp_mean_df.sort_values(by='happiness_score', ascending=False)
display(merged_data_sorted)
```

```python
from scipy.stats import iqr
Asia = happyness_df.groupby('region').get_group('Asia')

happiness_Q1 = Asia.describe().T.loc['happiness_score', '25%']
happiness_Q3 = Asia.describe().T.loc['happiness_score', '75%']
print(happiness_Q1, happiness_Q3)

iqr_value = iqr(Asia.happiness_score)
print(iqr_value)
lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')

outliers = Asia[(Asia['happiness_score'] < lower_bound) | (Asia['happiness_score'] > upper_bound)]
outliers
```

#### 2023년도 행복지수 보고서
- 대륙별로 묶어 행복지수를 기준으로 비교하였을때 높은 상관관계가 보이는 지표 3가지(1인당 국민소득,건강, 사회적 지원)를 기준으로 국가들의 표준편차를 구분

- 행복지수가 높은대륙은 사회적보장(1인당 국민소득,건강, 사회적 관계)등이 잘 갖춰져 있으며 대륙 내 국가별로 비슷한 수준을 이루고 있으며
행복지수가 낮은 대륙일수록 사회적 보장은 부실하며 개인간의 격차가 심화되어있다.



📌결론  
행복지수의 영향력이 높은 수치들인 개인의 경제력, 사회적 관계, 건강 등 자신 즉 국가 내 개인들의 비교가 쉬운 수치들이며 이들의 평균이 높고 편차가 낮을수록
행복지수는 높은 연관성을 보이며 국가간의 비교 즉 개인의 자유도, 국가의 부패정도는 개인의 행복도는 국가 내의 타인들과의 비교는 의미가 낮으며 행복지수의
유의미한 영향이 없다고 보인다.
